```python
import numpy as np #import modul numpy sebagai np
import cv2 #import modul cv2
import matplotlib.pyplot as plt #import modul matplotlib sebagai plt
```

```python
img = cv2.imread("D:\Semester 6\PCD\cod.jpg") #untuk membaca gambar yang ada pada folder semester 6 -> PCD -> cod.jpg dan menyimpannya pada variabel img
img_height = img.shape[0] #untuk mendapatakan informasi nilai tinggi dari gambar dan disimpan pada variabel img_height
img_width = img.shape[1] #untuk mendapatkan informasi nilai lebar dari gambar dan disimpan pada variabel img_width
img_channel = img.shape[2] #untuk mendapatkan informasi jumlah channel dari gambar dan disimpan pada variabel img_channel
img_type = img.dtype #untuk mendapatkan informasi tipe data dari gambar, seperti uint8 atau float32 dan disimpan pada variabel img_type
```

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_brightness
```

```python
def brighter(nilai): #mendefinisikan sebuah fungsi brighter dengan parameter nilai
    for y in range(0, img_height): #untuk melakukan looping pada setiap baris (y) dalam gambar
        for x in range(0, img_width): #untuk melakukan looping pada setiap kolom (x) dalam gambar
            red = img[y][x][0] #mengambil setiap pixel (x,y) nilai red pada gambar
            green = img[y][x][1] #mengambil setiap pixel (x,y) nilai green pada gambar
            blue = img[y][x][2] #mengambil setiap pixel (x,y) nilai blue pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray += nilai #menambahkan nilai pada nilai gray
            if gray > 255: #mengecek nilai keabuan jika lebih dari 255 
                gray = 255 #jika lebih dari 255, maka nilai keabuan diset jadi 255
            if gray < 0: #mengecek nilai keabuan jika kurang dari 0
                gray = 0 #jika kurang dari 0, maka nilai keabuan diset jadi 0
            img_brightness[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (x,y) disimpan pada variabel img_brightness
```

```python
brighter(-100) #mengatur nilai brighter jadi -100
plt.imshow(img_brightness) #menampilkan gambar pada variabel img_brightness
plt.title("Brightness -100") #diberi judul Brightness -100
plt.show() #untuk menampilkan output di layar

brighter(100) #mengatur nilai brighter jadi 100
plt.imshow(img_brightness) #menampilkan gambar pada variabel img_brightness
plt.title("Brightness 100") #diberi judul Brightness 100
plt.show() #untuk menampilkan output di layar
```

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_rgbbrightness
```

```python
def rgbbrighter(nilai): #mendefinisikan sebuah fungsi rgbbrighter dengan parameter nilai
    for y in range(0, img_height):  #untuk melakukan looping pada setiap baris (y) dalam gambar
        for x in range(0, img_width): #untuk melakukan looping pada setiap kolom (x) dalam gambar.
            red = img[y][x][0] #mengambil setiap pixel (x,y) nilai red pada gambar
            red += nilai #menambakan nilai pada nilai red
            if red > 255: #mengecek nilai red jika lebih dari 255
                red = 255 #jika lebih dari 255, maka nilai red diset jadi 255
            if red < 0: #mengecek nilai red jika kurang dari 0
                red = 0 #jika kurang dari 0, maka nilai red diset jadi 0
            green = img[y][x][1] #mengambil setiap pixel (x,y) nilai green pada gambar
            green += nilai #menambahkan nilai pada nilai green
            if green > 255: #mengecek nilai green jika lebih dari 255
                green = 255 #jika lebih dari 255, maka nilai green diset jadi 255
            if green < 0: #mengecek nilai green jika kurang dari 0
                green = 0 #jika kurang dari 0, maka nilai green diset jadi 0
            blue = img[y][x][2] #mengambil setiap pixel(x,y) nilai blue pada gambar
            blue += nilai #menambahkan nilai pada nilai blue
            if blue > 255: #mengecek nilai blue jika lebih dari 255
                blue = 255 #jika lebih dari 255, maka nilai blue diset jadi 255
            if blue < 0: #mengecek nilai blue jika kurang dari 0
                blue = 0 #jika kurang dari 0, maka nilai blue diset jadi 0
            img_rgbbrightness[y][x] = (red, green, blue) #nilai red, green, blue yang telah diubah pada setiap pixel (x,y) disimpan pada variabel img_rgbbrightness
```

```python
rgbbrighter(-100) #mengatur nilai rgbbrighter jadi -100
plt.imshow(img_rgbbrightness) #menampilkan gambar pada variabel img_rgbbrightness
plt.title("Brightness -100") #diberi judul Brightness -100
plt.show() #untuk menampilkan output di layar

rgbbrighter(100) #mengatur nilai rgbbrighter jadi 100
plt.imshow(img_rgbbrightness) #menampilkan gambar pada variabel img_rgbbrightness
plt.title("Brightness 100") #diberi judul Brightness 100
plt.show() #untuk menampilkan output dilayar
```

```python
img_contrass = np.zeros(img.shape, dtype=np.uint8) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_contrass
```

```python
def contrass(nilai): #mendefinisikan sebuah fungsi contrass dengan parameter nilai
    for y in range(0, img_height): #untuk melakukan looping pada setiap baris (y) dalam gambar
        for x in range(0, img_width):  #untuk melakukan looping pada setiap kolom (x) dalam gambar
            red = img[y][x][0] #mengambil setiap pixel (x,y) nilai red pada gambar
            green = img[y][x][1] #mengambil setiap pixel (x,y) nilai green pada gambar
            blue = img[y][x][2] #mengambil setiap pixel (x,y) nilai blue pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray *= nilai #mengalikan nilai gray 
            if gray > 255: #mengecek nilai gray jika lebih dari 255
                gray = 255 #jika lebih dari 255, maka nilai gray diset jadi 255
            img_contrass[y][x] = (gray, gray, gray) #menyimpan nilai keabuan yang telah diubah pada setiap pixel (x,y) dalam variabel img_contrast
```

```python
contrass(2) #mengatur nilai contrass jadi 2
plt.imshow(img_contrass) #menampilkan gambar pada variabel img_contrass
plt.title("Contrass 2") #diberi judul Contrass 2
plt.show() #untuk menampilkan output di layar

contrass(3) #mengatur nilai contrass jadi 3
plt.imshow(img_contrass) #menampilkan gambar pada variabel img_contrass
plt.title("Contrass 3") #diberi judul Contrass 3
plt.show() #untuk menampilkan output di layar
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)  #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_autocontrass
```

```python
def autocontrass(): #mendefinisikan fungsi autocontrass
    xmax = 300 #inisialisasi variabel xmax dengan nilai 300 
    xmin = 0 #inisialisasi variabel xmin dengan nilai 0
    d = 0 #inisialisasi variabel d dengan nilai 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height): #untuk melakukan looping pada setiap baris (y) dalam gambar
        for x in range(0, img_width): #untuk melakukan looping pada setiap kolom (x) dalam gambar
            red = img[y][x][0] #mengambil setiap pixel (x,y) nilai red pada gambar
            green = img[y][x][1] #mengambil setiap pixel (x,y) nilai green pada gambar
            blue = img[y][x][2] #mengambil setiap pixel (x,y) nilai blue pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            if gray < xmax: #mengecek nilai gray kurang dari nilai xmax atau tidak
                xmax = gray #jika kurang dari nilai xmax, maka nilai xmax diset jadi nilai gray
            if gray > xmin: #mengecek nilai gray lebih dari xmin atau tidak
                xmin = gray #jika lebih dari xmin, maka nilai xmin diset jadi nilai gray
    d = xmin-xmax #operasi pengurangan untuk mengetahui selisih dari nilai xmin dan xmax
    for y in range(0, img_height): #untuk melakukan looping pada setiap baris (y) dalam gambar
        for x in range(0, img_width): #untuk melakukan looping pada setiap kolom (x) dalam gambar
            red = img[y][x][0] #mengambil setiap pixel (x,y) nilai red pada gambar
            green = img[y][x][1] #mengambil setiap pixel (x,y) nilai green pada gambar
            blue = img[y][x][2] #mengambil setiap pixel (x,y) nilai blue pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray = int(float(255/d) * (gray-xmax)) #menghitung nilai gray dengan menggunakan rumus
            img_autocontrass[y][x] = (gray, gray, gray) #menyimpan nilai keabuan yang telah dihitung pada setiap pixel (x, y) dalam variabel img_autocontrass
```

```python
autocontrass() #memanggil fungsi autocontrass
plt.imshow(img_autocontrass) #menampilkan gambar pada variabel img_autocontrass
plt.title("Contrass Autolevel") #diberi judul Contrass Autolevel
plt.show() #untuk menampilkan output pada layar
```
